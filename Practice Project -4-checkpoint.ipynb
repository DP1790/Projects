# Abalone Case Study

The age of abalone is determined by cutting the shell through the cone, staining it, and counting the number of rings through a microscope. Given is the attribute name, attribute type, the measurement unit and a brief description. The number of rings is the value to predict. Hence the target variable/class is multiclasssification & we need to predict the rings of each abalone which will lead us to the age of that abalone.

# Removing outliers using zscore

```python
from scipy.stats import zscore
z = np.abs(zscore(df))
threshold = 3
x_new = df[(z<3).all(axis = 1)]
```

```python
print ("shape before and after")
print ("shape befor" .ljust(20), ":", df.shape)
print ("shape after" .ljust(20), ":", x_new.shape)
print ("Percentage loss" .ljust(20), ":", (df.shape[0]-x_new.shape[0])/df.shape[0])
```

```python
# Lets check again if outliers are removed
plt.figure(figsize=(20,15))
graph = 1

for column in x_new:
    if graph<=12:
        ax = plt.subplot(4,3,graph)
        sns.boxplot(x_new[column], orient='v')
        plt.xlabel(column,fontsize=12)
    
    graph+=1
plt.show()
```

# As we can see now, the data is not skewed & outliers is also removed

We have seen in the heatmap that there are few columns, which is having high multicolinearity, Lets split the data & check how much class is colinear with each features and accordinly we can drop one of the features if required

```python
## Divide data set into features and label

y = x_new['Rings']
x = x_new.drop(columns=['Rings'])
```

```python
## Visualizing the relationship

plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in x_new:
    if plotnumber<=8:
        ax = plt.subplot(4,4,plotnumber)
        sns.scatterplot(x_new[column],y)
        plt.xlabel(column,fontsize=20)
        plt.ylabel('RINGS', fontsize =20)
    
    plotnumber+=1
plt.tight_layout()
```

All the independent variables is having some relation with class. Hence no need to drop any feature.

# Creating train test split

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=234)
```

```python
## DecisionTreeclassifier

dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
preddt =dt.predict(x_test)
print(accuracy_score(y_test, preddt))
print(confusion_matrix(y_test, preddt))
print(classification_report(y_test, preddt))
```

```python
## Random Forest Classifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
predrf =rf.predict(x_test)
print(accuracy_score(y_test, predrf))
print(confusion_matrix(y_test, predrf))
print(classification_report(y_test, predrf))
```

```python
# KNN Classifier
kn = KNeighborsClassifier()
kn.fit(x_train, y_train)
predkn =kn.predict(x_test)
print(accuracy_score(y_test, predkn))
print(confusion_matrix(y_test, predkn))
print(classification_report(y_test, predkn))
```

```python
## SVC
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train, y_train)
predsvc =svc.predict(x_test)
print(accuracy_score(y_test, predsvc))
print(confusion_matrix(y_test, predsvc))
print(classification_report(y_test, predsvc))
```

# Abalone Case Study

The age of abalone is determined by cutting the shell through the cone, staining it, and counting the number of rings through a microscope. Given is the attribute name, attribute type, the measurement unit and a brief description. The number of rings is the value to predict. Hence we need to add 1.5 to get the age of the abalone, which is target variable/class & it is regression problem as the target variable is continous.

```python
## Importing libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, roc_auc_score, classification_report, plot_roc_curve
import statsmodels.api as sm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
import pickle
import warnings
warnings.filterwarnings('ignore')
```

# Removing outliers using zscore

```python
from scipy.stats import zscore
z = np.abs(zscore(df))
threshold = 3
x_new = df[(z<3).all(axis = 1)]
```

```python
print ("shape before and after")
print ("shape befor" .ljust(20), ":", df.shape)
print ("shape after" .ljust(20), ":", x_new.shape)
print ("Percentage loss" .ljust(20), ":", (df.shape[0]-x_new.shape[0])/df.shape[0])
```

```python
# Lets check again if outliers are removed
plt.figure(figsize=(20,15))
graph = 1

for column in x_new:
    if graph<=12:
        ax = plt.subplot(4,3,graph)
        sns.boxplot(x_new[column], orient='v')
        plt.xlabel(column,fontsize=12)
    
    graph+=1
plt.show()
```

```python
x_new.skew()
```

```python
# lets plot again in distribution plot to visualize
plt.figure(figsize=(15,30))
plotnumber = 1

for column in x_new:
    if plotnumber<=10:
        ax = plt.subplot(4,3,plotnumber)
        sns.distplot(x_new[column])
        plt.xlabel(column,fontsize=20)
    
    plotnumber+=1
plt.show()
```

# As we can see now, the data is not skewed & outliers is also removed

We have seen in the heatmap that there are few columns, which is having high multicolinearity, Lets split the data & check how much class is colinear with each features and accordinly we can drop one of the features if required

```python
## Divide data set into features and label

y = x_new['age']
x = x_new.drop(columns=['age'])
```

```python
## Visualizing the relationship

plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in x:
    if plotnumber<=8:
        ax = plt.subplot(4,4,plotnumber)
        sns.scatterplot(x[column],y)
        plt.xlabel(column,fontsize=20)
        plt.ylabel('age', fontsize =20)
    
    plotnumber+=1
plt.tight_layout()
```

# Creating train test split

```python
# Split the data into train and test. Model will be built on training data and tested on test data.
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=320)
y_train.head()
```

# Model initiating and training

```python
regression = LinearRegression()
regression.fit(x_train,y_train)
```

# Predict the Age

```python
y_pred = regression.predict(x_test)
y_pred
```

```python
## adjusted R2 score
regression.score(x_train,y_train)
```

```python
regression.score(x_test,y_test)
```

```python
# Lets visualize the Actual Vs Prediction
plt.scatter(y_test,y_pred)
plt.xlabel('Actual Age')
plt.ylabel('Aredicted Age')
plt.title('Actual Vs Prediction')
plt.show()
```

Lets check with regularization if the model is overfitted or not

```python
# LassoCV
from sklearn.linear_model import Ridge, Lasso, RidgeCV, LassoCV
```

```python
# best alpha parameter
alpha = lasscv.alpha_
alpha
```

```python
# now that we have the best parameter, lets use Lasso regression and see how well our data has fitted before

lasso_reg = Lasso(alpha)
lasso_reg.fit(x_train, y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

```python
## using RidgeCV

ridgecv = RidgeCV(alphas= np.arange(0.001,0.01,0.1), normalize=True)
ridgecv.fit(x_train, y_train)
```

```python
ridgecv.alpha_
```

```python
ridge_model = Ridge(alpha =ridgecv.alpha_)
ridge_model.fit(x_train, y_train)
```

```python
ridge_model.score(x_test,y_test)
```

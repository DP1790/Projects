## Data description

```python
## Importing libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
## Read the CSV file and convert into dataframe

df = pd.read_csv('happiness_score_dataset.csv')
df.head()
```

```python
df.shape ## checking the data dimensions
```

We have 158 rows & 12 columns including the target column.

```python
df.isnull().sum() ## checking the null values
```

```python
## understanding the data at high level. checking the statistics of data set
df.describe()
```

```python
## dropping the unwanted column

data = df.drop(columns=['Country','Region','Happiness Rank'])
data.head()
```

Lets visualize the data & analyze the relationship between independent and dependent variables.

```python
## lets see how data is distributed for every column

plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in data:
    if plotnumber<=9:
        ax = plt.subplot(3,3,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=20)
    
    plotnumber+=1
plt.tight_layout()
```

```python
data.skew()
```

The data distribution looks decent enough & there doesn't seem to be any skewness, only standard error has little skewness which can be considerable. So we can proceed !

Let's observe the relationship between independent variable & dependent variable.

```python
## Divide data set into features and label

y = data['Happiness Score']
x = data.drop(columns=['Happiness Score'])
```

```python
y
```

```python
x
```

```python
## Visualizing the relationship

plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in x:
    if plotnumber<=8:
        ax = plt.subplot(4,4,plotnumber)
        sns.scatterplot(x[column],y)
        plt.xlabel(column,fontsize=20)
        plt.ylabel('Happiness Score', fontsize =20)
    
    plotnumber+=1
plt.tight_layout()
```

The relationship between independent variable & dependent variable look fairly linear except standard error. Hence we need to drop the feature.
The data looks scaled. Hence we can go ahead with model building

```python
## droping the standard eror column

x = data.drop(columns=['Standard Error','Happiness Score'])
y = data['Happiness Score']
```

```python
x.head(2)
```

The data looks scaled. Hence we can go ahead and use linear regression & see how good it fits. but we need to split our data in train and test.

```python
# Split the data into train and test. Model will be built on training data and tested on test data.
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=320)
y_train.head()
```

## Model initiating and training

```python
regression = LinearRegression()
regression.fit(x_train,y_train)
```

## Predict the Happiness Score

```python
y_pred = regression.predict(x_test)
y_pred
```

```python
## adjusted R2 score
regression.score(x_train,y_train)
```

```python
regression.score(x_test,y_test)
```

Lest plot and visualize

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Actual Happiness Score')
plt.ylabel('predicted Happiness Score')
plt.title('Actual Vs Prediction')
plt.show()
```

## Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
y_pred = regression.predict(x_test)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

Lets check with regularization if the model is overfitted or not

```python
# LassoCV
from sklearn.linear_model import Ridge, Lasso, RidgeCV, LassoCV
```

```python
lasscv = LassoCV(alphas = None, max_iter= 1000, normalize=True)
lasscv.fit(x_train, y_train)
```

```python
# best alpha parameter
alpha = lasscv.alpha_
alpha
```

```python
# now that we have the best parameter, lets use Lasso regression and see how well our data has fitted before

lasso_reg = Lasso(alpha)
lasso_reg.fit(x_train, y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

```python
## using RidgeCV

ridgecv = RidgeCV(alphas= np.arange(0.001,0.01,0.1), normalize=True)
ridgecv.fit(x_train, y_train)
```

```python
ridgecv.alpha_
```

```python
ridge_model = Ridge(alpha =ridgecv.alpha_)
ridge_model.fit(x_train, y_train)
```

```python
ridge_model.score(x_test,y_test)
```

Hence we could conclude that the linear regression model is fine & not overfitted.

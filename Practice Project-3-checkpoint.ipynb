```python
# lets plot distribution plot to visualize
plt.figure(figsize=(15,30))
plotnumber = 1

for column in x:
    if plotnumber<=12:
        ax = plt.subplot(3,4,plotnumber)
        sns.distplot(x[column])
        plt.xlabel(column,fontsize=20)
    
    plotnumber+=1
plt.show()
```

```python
# lets visualize the outliers using box plot
plt.figure(figsize=(20,15))
graph = 1

for column in x:
    if graph<=11:
        ax = plt.subplot(4,3,graph)
        sns.boxplot(x[column], orient='v')
        plt.xlabel(column,fontsize=12)
    
    graph+=1
plt.show()
```

# Removing outliers using IQR

```python
# find the IQR to identify outliers

# 1st quartile
q1  = x.quantile(0.25)

#3rd quartile
q3 = x.quantile(0.75)

#IQR
iqr = q3-q1
```

```python
# Removing the higher outliers

x_new = x[~((x < (q1 - 1.5 * iqr)) | (x > (q3 + 1.5 * iqr))).any(axis = 1)]
```

```python
print ("shape before and after")
print ("shape befor" .ljust(20), ":", x.shape)
print ("shape after" .ljust(20), ":", x_new.shape)
print ("Percentage loss" .ljust(20), ":", (x.shape[0]-x_new.shape[0])/x.shape[0])
```

```python
plt.figure(figsize=(20,15))
graph = 1

for column in x_new:
    if graph<=11:
        ax = plt.subplot(4,3,graph)
        sns.boxplot(x_new[column], orient='v')
        plt.xlabel(column,fontsize=12)
    
    graph+=1
plt.show()
```

# Removing outliers using zscore

```python
from scipy.stats import zscore
z = np.abs(zscore(x))
threshold = 3
x_new1 = x[(z<3).all(axis = 1)]
```

```python
print ("shape before and after")
print ("shape befor" .ljust(20), ":", x.shape)
print ("shape after" .ljust(20), ":", x_new1.shape)
print ("Percentage loss" .ljust(20), ":", (x.shape[0]-x_new1.shape[0])/x.shape[0])
```

```python
plt.figure(figsize=(20,15))
graph = 1

for column in x_new1:
    if graph<=11:
        ax = plt.subplot(4,3,graph)
        sns.boxplot(x_new1[column], orient='v')
        plt.xlabel(column,fontsize=12)
    
    graph+=1
plt.show()
```

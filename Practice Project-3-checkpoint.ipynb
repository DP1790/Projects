```python
## lets check if there is any outliers
df.boxplot(figsize=(20,8))
plt.subplots_adjust(bottom=0.25)
plt.show()
```

```python
# lets plot distribution plot to visualize
plt.figure(figsize=(15,30))
plotnumber = 1

for column in df:
    if plotnumber<=12:
        ax = plt.subplot(3,4,plotnumber)
        sns.distplot(df[column])
        plt.xlabel(column,fontsize=20)
    
    plotnumber+=1
plt.show()
```

```python
# lets visualize the outliers using box plot
plt.figure(figsize=(20,15))
graph = 1

for column in df:
    if graph<=12:
        ax = plt.subplot(4,3,graph)
        sns.boxplot(df[column], orient='v')
        plt.xlabel(column,fontsize=12)
    
    graph+=1
plt.show()
```

# Removing outliers using IQR

```python
# find the IQR to identify outliers

# 1st quartile
q1  = df.quantile(0.25)

#3rd quartile
q3 = df.quantile(0.75)

#IQR
iqr = q3-q1
```

```python
# Removing the outliers

x_new = df[~((df < (q1 - 1.5 * iqr)) | (df > (q3 + 1.5 * iqr))).any(axis = 1)]
```

```python
print ("shape before and after")
print ("shape befor" .ljust(20), ":", df.shape)
print ("shape after" .ljust(20), ":", x_new.shape)
print ("Percentage loss" .ljust(20), ":", (df.shape[0]-x_new.shape[0])/df.shape[0])
```

```python
# Lets check again if outliers are removed
plt.figure(figsize=(20,15))
graph = 1

for column in x_new:
    if graph<=12:
        ax = plt.subplot(4,3,graph)
        sns.boxplot(x_new[column], orient='v')
        plt.xlabel(column,fontsize=12)
    
    graph+=1
plt.show()
```

# Removing outliers using zscore

```python
from scipy.stats import zscore
z = np.abs(zscore(df))
threshold = 3
x_new1 = df[(z<3).all(axis = 1)]
```

```python
print ("shape before and after")
print ("shape befor" .ljust(20), ":", df.shape)
print ("shape after" .ljust(20), ":", x_new1.shape)
print ("Percentage loss" .ljust(20), ":", (df.shape[0]-x_new1.shape[0])/df.shape[0])
```

```python
# Lets check again if outliers are removed
plt.figure(figsize=(20,15))
graph = 1

for column in x_new1:
    if graph<=12:
        ax = plt.subplot(4,3,graph)
        sns.boxplot(x_new1[column], orient='v')
        plt.xlabel(column,fontsize=12)
    
    graph+=1
plt.show()
```

After using both the methods we can see we lose more % of data when we use the IQR, hence we should consider the zscore

```python
## splitting the independent and target variables in x & y
x = x_new1.drop("quality",axis = 1)
y = x_new1["quality"]
```

Lets check for skewness

```python
x.skew()
```

```python
# Using Power transformation to remove the skewness
from sklearn.preprocessing import power_transform
data = power_transform(x)
data = pd.DataFrame(data, columns = x.columns)
```

Hence we have removed the outliers as well as skewness. We can proceed with the model building

# Finding best Random_ state

```python
maxAccu = 0
maxRS = 0
for i in range(1,250):
    x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = .25, random_state= i)
    rf= RandomForestClassifier()
    rf.fit(x_train,y_train)
    predrf = rf.predict(x_test)
    acc = accuracy_score(y_test, predrf)
    if acc > maxAccu:
        maxAccu = acc
        maxRS = i
        
print("Best accuracy is ", maxAccu, "on Random_state", maxRS)
        
```

We have found the best random state as 231 & we will create the train test split using the same random state(231)

# Creating train test splt

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=maxRS)
```

```python
## DecisionTreeclassifier

dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
preddt =dt.predict(x_test)
print(accuracy_score(y_test, preddt))
print(confusion_matrix(y_test, preddt))
print(classification_report(y_test, preddt))

```

```python
## Random Forest Classifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
predrf =rf.predict(x_test)
print(accuracy_score(y_test, predrf))
print(confusion_matrix(y_test, predrf))
print(classification_report(y_test, predrf))
```

```python
# KNN Classifier
kn = KNeighborsClassifier()
kn.fit(x_train, y_train)
predkn =kn.predict(x_test)
print(accuracy_score(y_test, predkn))
print(confusion_matrix(y_test, predkn))
print(classification_report(y_test, predkn))
```

```python
## SVC
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train, y_train)
predsvc =svc.predict(x_test)
print(accuracy_score(y_test, predsvc))
print(confusion_matrix(y_test, predsvc))
print(classification_report(y_test, predsvc))
```

We are getting highest accuracy with RandomForest(77%) but it can due to overfitting, so we need to check the cross validation scores

```python
from sklearn.model_selection import cross_val_score

scr = cross_val_score(dt, x, y, cv=5)
print("Cross validation score for Decision Tree model:" , scr.mean())
```

```python
scr = cross_val_score(rf, x, y, cv=5)
print("Cross validation score for Random Forest model:" , scr.mean())
```

```python
scr = cross_val_score(kn, x, y, cv=5)
print("Cross validation score for KNN model:" , scr.mean())
```

```python
scr = cross_val_score(svc, x, y, cv=5)
print("Cross validation score for SVC model:" , scr.mean())
```

Although we have highest accuracy with RandomForest(77%) but based on Cross Validation we have best fit model is KNN with 64% accuracy score

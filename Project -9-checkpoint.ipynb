# Flight Price Prediction

##### Problem Statement:
- Flight ticket prices can be something hard to guess, today we might see a price, check out the price of the same flight tomorrow, it will be a different story. We might have often heard travellers saying that flight ticket prices are so unpredictable. Here you will be provided with prices of flight tickets for various airlines between the months of March and June of 2019 and between various cities.

- Size of training set: 10683 records
- Size of test set: 2671 records

##### Target Variable:
- build a machine learning model to predict the price of the flight ticket.

##### Finding best Random_ state

```python
from sklearn.ensemble import RandomForestRegressor
maxr2score = 0
maxRS = 0

for i in range (1, 150):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = i)
    RF = RandomForestRegressor()
    RF.fit(x_train, y_train)
    predrf = RF.predict(x_test)
    r2 = r2_score(y_test, predrf)
    if r2>maxr2score:
        maxr2score = r2
        maxRS = i
print ('Max R2 score obtained is', maxr2score, 'on Random State', maxRS)
```

##### Creating train test split

```python
# Split the data into train and test. Model will be built on training data and tested on test data.
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=maxRS)
y_train.head()
```

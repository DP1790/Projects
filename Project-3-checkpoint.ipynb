# Baseball Case Study 

This dataset utilizes data from 2014 Major League Baseball seasons in order to develop an algorithm that predicts the number of wins for a given team in the 2015 season based on several different indicators of success. There are 16 different features that will be used as the inputs to the machine learning and the output will be a value that represents the number of wins.


Input features: Runs, At Bats, Hits, Doubles, Triples, Homeruns, Walks, Strikeouts, Stolen Bases, Runs Allowed, Earned Runs, Earned Run Average (ERA), Shutouts, Saves, Complete Games and Errors

Output: Number of predicted wins (W)

# Lets check for outliers & skewness.

```python
# lets visualize the outliers using box plot
plt.figure(figsize=(17,9))
graph = 1

for column in df:
    if graph<=17:
        ax = plt.subplot(6,3,graph)
        sns.boxplot(df[column], orient='h')
        plt.xlabel(column,fontsize=11)
    
    graph+=1
plt.show()
```

There are few columns which are having outliers.

# Removing outliers using zscore

```python
from scipy.stats import zscore
z = np.abs(zscore(df))
threshold = 3
df1 = df[(z<3).all(axis = 1)]
```

```python
print ("shape before and after")
print ("shape befor" .ljust(20), ":", df.shape)
print ("shape after" .ljust(20), ":", df1.shape)
print ("Percentage loss" .ljust(20), ":", (df.shape[0]-df1.shape[0])/df.shape[0]*100)
```

# Removing outliers using IQR

```python
# find the IQR to identify outliers

# 1st quartile
q1  = df.quantile(0.25)

#3rd quartile
q3 = df.quantile(0.75)

#IQR
iqr = q3-q1
```

```python
# Removing the outliers

df2 = df[~((df < (q1 - 1.5 * iqr)) | (df > (q3 + 1.5 * iqr))).any(axis = 1)]
```

```python
print ("shape before and after")
print ("shape befor" .ljust(20), ":", df.shape)
print ("shape after" .ljust(20), ":", df2.shape)
print ("Percentage loss" .ljust(20), ":", (df.shape[0]-df2.shape[0])/df.shape[0]*100)
```

Using IQR, we are losing more data, so lets consider zscore & proceed for checking the skewness

```python
# lets plot distribution plot to visualize
plt.figure(figsize=(15,30))
plotnumber = 1

for column in df1:
    if plotnumber<=17:
        ax = plt.subplot(6,3,plotnumber)
        sns.distplot(df1[column])
        plt.xlabel(column,fontsize=20)
    
    plotnumber+=1
plt.show()
```

```python
# Lets check the skewness
df1.skew()
```

We have skewness in column H, CG, SHO, SV & E as compare to other features.

```python
# Using Power transformation to remove the skewness
from sklearn.preprocessing import power_transform
df1[['CG','SHO','SV','E']]=power_transform(
    df1[['CG','SHO','SV','E']],method='yeo-johnson')
```

```python
df1.skew()
```

```python
# After using power transformation for H column, it has transformed the all values to 0. Hence using to log transformation to reduce the skewness
df1['H'] = np.log1p(df1['H'])
```

```python
## Divide data set into features and label
y = df1['W']
x = df1.drop(columns=['W'], axis=1)
```

# Finding best Random_ state

```python
maxr2score = 0
maxRS = 0

for i in range (1, 150):
    x_train, x_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.22, random_state = i)
    LR = LinearRegression()
    LR.fit(x_train, y_train)
    predlr = LR.predict(x_test)
    r2 = r2_score(y_test, predlr)
    if r2>maxr2score:
        maxr2score = r2
        maxRS = i
print ('Max R2 score obtained is', maxr2score, 'on Random State', maxRS)
```

# Creating train test split

```python
# Split the data into train and test. Model will be built on training data and tested on test data.
x_train,x_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.22, random_state=maxRS)
y_train.head()
```
